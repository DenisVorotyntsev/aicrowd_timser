# Description

# Config 

```python
PATH_TO_ACTUAL_DATA = "../data/external/val.csv"
PATH_TO_PREDICTION = "../output/prediction.csv"
```

# Libs 

```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt 
import seaborn as sns
sns.set()
```

# Code

```python
y_true = pd.read_csv(PATH_TO_ACTUAL_DATA)
y_hat = pd.read_csv(PATH_TO_PREDICTION)
```

```python
plt.plot(y_true["value"], label="actual data")
plt.plot(y_hat["value"], label="prediction")
plt.legend()
plt.show()
```
